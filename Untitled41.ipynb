```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import torch
import torch.nn as nn
import torch.optim as optim
from scipy.stats import gaussian_kde
```

```python
df = pd.read_csv("/content/india-air-quality-data.csv", encoding="latin1")
x = df["no2"].dropna().values

z = x + a_r * np.sin(b_r * x)
```

```python
scaler = MinMaxScaler()
z = z.reshape(-1,1)
z_scaled = scaler.fit_transform(z)
```

```python

class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(1,32),
            nn.ReLU(),
            nn.Linear(32,32),
            nn.ReLU(),
            nn.Linear(32,1)
        )

    def forward(self,x):
        return self.model(x)


class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(1,32),
            nn.LeakyReLU(0.2),
            nn.Linear(32,32),
            nn.LeakyReLU(0.2),
            nn.Linear(32,1),
            nn.Sigmoid()
        )

    def forward(self,x):
        return self.model(x)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

G = Generator().to(device)
D = Discriminator().to(device)

criterion = nn.BCELoss()

optimizer_G = optim.Adam(G.parameters(), lr=0.0002)
optimizer_D = optim.Adam(D.parameters(), lr=0.0002)

z_tensor = torch.tensor(z_scaled, dtype=torch.float32).to(device)

epochs = 2000
batch_size = 128
```

```python
for epoch in range(epochs):

    idx = np.random.randint(0, z_tensor.shape[0], batch_size)
    real = z_tensor[idx]

    real_labels = torch.ones(batch_size,1).to(device)
    fake_labels = torch.zeros(batch_size,1).to(device)

    # Train Discriminator
    noise = torch.randn(batch_size,1).to(device)
    fake = G(noise)

    D_real = D(real)
    D_fake = D(fake.detach())

    loss_real = criterion(D_real, real_labels)
    loss_fake = criterion(D_fake, fake_labels)

    loss_D = loss_real + loss_fake

    optimizer_D.zero_grad()
    loss_D.backward()
    optimizer_D.step()

    # Train Generator
    noise = torch.randn(batch_size,1).to(device)
    fake = G(noise)

    output = D(fake)

    loss_G = criterion(output, real_labels)

    optimizer_G.zero_grad()
    loss_G.backward()
    optimizer_G.step()

    if epoch % 200 == 0:
        print(f"Epoch {epoch} | D Loss: {loss_D.item()} | G Loss: {loss_G.item()}")
```
